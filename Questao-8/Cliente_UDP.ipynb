```python
import socket
```

```python
def cliente_udp():
    UDP_IP = "127.0.0.1"
    UDP_PORT = 5005

    sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    valor = input("Digite o valor a ser convertido (em R$): ")
    moeda_destino = input("Digite a moeda de destino: ")

    mensagem = f"{valor},{moeda_destino}"
    sock.sendto(mensagem.encode(), (UDP_IP, UDP_PORT))

    data, _ = sock.recvfrom(1024)
    print("Resultado da conversão:", data.decode())

cliente_udp()
```
